```python
#Ejemplos de variables y tipos basicos
entero = 10                 #int
numero_real = 3.1416        #float
texto = 'Ada Lovelace'      #str
es_activo = True            #bool
nada = None                 #NoneType

print(f"Variable entero con valor {entero} es de tipo {type(entero).__name__}")
print(f"Variable numero real con valor {numero_real} es de tipo {type(numero_real).__name__}")
print(f"Variable texto con valor {texto} es de tipo {type(texto).__name__}")
print(f"Variable es activo con valor {es_activo} es de tipo {type(es_activo).__name__}")
print(f"Variable nada con valor {nada} es de tipo {type(nada).__name__}")
```
